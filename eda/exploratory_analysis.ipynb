## Exploratory Data Analysis (EDA)

```python
#import packages 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Ensure that visualizations are displayed in the notebook
%matplotlib inline
```

```python
# Load the synthetic data
df = pd.read_csv('../data_preparation/synthetic_loan_data.csv')

# Display the first few rows of the dataset
df.head()
```

### Summary Statistics

```python
df.describe()
```

```python
# Distribution of Loan Amounts 
# Histogram for Last_Loan_Amount
sns.histplot(df['Last_Loan_Amount'], kde=True)
plt.title('Distribution of Last Loan Amounts')
plt.show()

```

```python
#Distribution of Loan Durations
# Histogram for Last_Loan_Duration
# Histogram for Last_Loan_Duration
sns.histplot(df['Last_Loan_Duration'], kde=True)
plt.title('Distribution of Last Loan Durations')
plt.show()
```

```python
#Defaulted vs. Non-Defaulted Loans
# Bar plot for Defaulted
# Bar plot for Defaulted
sns.countplot(x='Defaulted', data=df, palette=["#FF9999", "#66B2FF"])
plt.title('Count of Defaulted vs Non-Defaulted Loans')
plt.show()
```

```python
# enhanced data generated 

# Load the synthetic data
df = pd.read_csv('../data_preparation/synthetic_loan_data_enhanced.csv')

# Display the first few rows of the dataset
df.head()

# Histogram for Last_Loan_Amount
sns.histplot(df['Last_Loan_Amount'], kde=True)
plt.title('Distribution of Last Loan Amounts')
plt.show()

# Histogram for Last_Loan_Duration_Weeks
sns.histplot(df['Last_Loan_Duration_Weeks'], kde=True)
plt.title('Distribution of Last Loan Durations (in weeks)')
plt.show()

# Histogram for Last_Loan_Term_Weeks
sns.histplot(df['Last_Loan_Term_Weeks'], kde=True)
plt.title('Distribution of Last Loan Terms (in weeks)')
plt.show()

# Bar plot for Defaulted
sns.countplot(x='Defaulted', data=df, palette=["#FF9999", "#66B2FF"])
plt.title('Count of Defaulted vs Non-Defaulted Loans')
plt.show()
```
